```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

print("Libraries imported successfully.")

```

```python
# Step 2: Create sales dataset

sales_data = {
    "Year": [2020, 2021, 2022, 2023, 2024],
    
    "Clarks_Revenue_Million_GBP": [950, 980, 1020, 994.5, 901.3],
    
    "Superdry_Revenue_Million_GBP": [704, 689, 610, 611, 580],
    
    "Clarks_Profit_Million_GBP": [-120, -85, 35.9, -39.8, -39.2],
    
    "Superdry_Profit_Million_GBP": [-30, -20, 15, -67.7, 50.5]
}

sales_df = pd.DataFrame(sales_data)

sales_df
```

```python
# Step 3: Calculate profit margin

sales_df["Clarks_Profit_Margin_%"] = (
    sales_df["Clarks_Profit_Million_GBP"] /
    sales_df["Clarks_Revenue_Million_GBP"]
) * 100

sales_df["Superdry_Profit_Margin_%"] = (
    sales_df["Superdry_Profit_Million_GBP"] /
    sales_df["Superdry_Revenue_Million_GBP"]
) * 100

sales_df
```

```python
# Step 4: Create demographics dataset

demographics_data = {
    "Brand": ["Clarks", "Clarks", "Clarks", "Superdry", "Superdry", "Superdry"],
    
    "Age_Group": [
        "18-24", "25-34", "35-54",
        "18-24", "25-34", "35-54"
    ],
    
    "Male_%": [
        48, 50, 52,
        44, 44.15, 46
    ],
    
    "Female_%": [
        52, 50, 48,
        56, 55.85, 54
    ],
    
    "Avg_Income_GBP": [
        22000, 28000, 35000,
        24000, 30000, 37000
    ]
}

demographics_df = pd.DataFrame(demographics_data)

demographics_df
```

```python
# Step 5: Save dataset

sales_df.to_csv("clarks_superdry_sales_dataset.csv", index=False)

demographics_df.to_csv("clarks_superdry_demographics_dataset.csv", index=False)

print("Datasets saved successfully.")
```

```python
# Step 6: Plot revenue trends

plt.figure(figsize=(10,6))

plt.plot(sales_df["Year"], sales_df["Clarks_Revenue_Million_GBP"], marker='o')
plt.plot(sales_df["Year"], sales_df["Superdry_Revenue_Million_GBP"], marker='o')

plt.title("Revenue Trends (Clarks vs Superdry)")
plt.xlabel("Year")
plt.ylabel("Revenue (Million GBP)")
plt.legend(["Clarks", "Superdry"])

plt.show()
```

```python
# Step 7: Profit margin visualization

plt.figure(figsize=(10,6))

plt.plot(sales_df["Year"], sales_df["Clarks_Profit_Margin_%"], marker='o')
plt.plot(sales_df["Year"], sales_df["Superdry_Profit_Margin_%"], marker='o')

plt.title("Profit Margin Trends")
plt.xlabel("Year")
plt.ylabel("Profit Margin (%)")

plt.legend(["Clarks", "Superdry"])

plt.show()
```

```python
# Step 8: Demographics visualization

sns.barplot(
    data=demographics_df,
    x="Age_Group",
    y="Female_%",
    hue="Brand"
)

plt.title("Female Customer Demographics by Age Group")

plt.show()
```

```python
# Step 9: Merge datasets for analysis

combined_df = sales_df.copy()

combined_df["Brand"] = "Both"

combined_df
```
