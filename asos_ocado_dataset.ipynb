```python
# Step 2: Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots in notebook
%matplotlib inline

# Set style
sns.set(style="whitegrid")
```

```python
# Step 3: ASOS dataset

asos_data = {
    "Year": [2021, 2022, 2023, 2024, 2025],
    
    # Revenue in £ millions
    "Revenue (£m)": [3910, 3980, 3540, 2896, 2465],
    
    # Gross margin %
    "Gross Margin (%)": [48.5, 47.8, 46.0, 43.4, 47.1],
    
    # Net profit margin %
    "Net Margin (%)": [-1.8, -5.0, -10.2, -11.7, -12.0],
    
    # Demographics
    "Age 16-24 (%)": [42, 43, 44, 45, 46],
    "Age 25-34 (%)": [38, 37, 36, 35, 34],
    "Age 35-44 (%)": [12, 12, 12, 12, 12],
    "Age 45+ (%)": [8, 8, 8, 8, 8],
    
    "Female (%)": [60, 61, 62, 63, 64],
    "Male (%)": [40, 39, 38, 37, 36]
}

asos_df = pd.DataFrame(asos_data)

asos_df
```

```python
# Step 4: Ocado dataset

ocado_data = {
    "Year": [2021, 2022, 2023, 2024, 2025],
    
    # Revenue in £ millions
    "Revenue (£m)": [2500, 2700, 2830, 3200, 2827],
    
    # Gross margin %
    "Gross Margin (%)": [18.0, 17.5, 16.2, 14.8, 13.8],
    
    # Net margin %
    "Net Margin (%)": [-3.8, -34.6, -65.4, -96.3, -23.2],
    
    # Demographics
    "Age 18-34 (%)": [28, 29, 30, 31, 32],
    "Age 35-54 (%)": [45, 44, 43, 42, 41],
    "Age 55+ (%)": [27, 27, 27, 27, 27],
    
    "Female (%)": [52, 53, 54, 54, 55],
    "Male (%)": [48, 47, 46, 46, 45]
}

ocado_df = pd.DataFrame(ocado_data)

ocado_df
```

```python
# Step 5: Add company column

asos_df["Company"] = "ASOS"
ocado_df["Company"] = "Ocado"

combined_df = pd.concat([asos_df, ocado_df], ignore_index=True)

combined_df
```

```python
# Step 6: Save to Excel

combined_df.to_excel("asos_ocado_dataset.xlsx", index=False)

print("Dataset saved successfully!")
```

```python
# Step 7: Plot revenue comparison

plt.figure(figsize=(10,6))

sns.lineplot(data=combined_df,
             x="Year",
             y="Revenue (£m)",
             hue="Company",
             marker="o")

plt.title("ASOS vs Ocado Revenue Comparison")
plt.ylabel("Revenue (£ millions)")
plt.show()
```

```python
# Step 8: Profit margin comparison

plt.figure(figsize=(10,6))

sns.lineplot(data=combined_df,
             x="Year",
             y="Gross Margin (%)",
             hue="Company",
             marker="o")

plt.title("Gross Profit Margin Comparison")
plt.ylabel("Gross Margin (%)")
plt.show()
```

```python
# Step 9: ASOS gender demographics

gender_data = asos_df[["Year", "Female (%)", "Male (%)"]]

gender_data.set_index("Year").plot(kind="bar", figsize=(10,6))

plt.title("ASOS Gender Demographics")
plt.ylabel("Percentage")
plt.show()
```

```python
# Step 10: Average metrics

summary = combined_df.groupby("Company").agg({

    "Revenue (£m)": "mean",
    "Gross Margin (%)": "mean",
    "Net Margin (%)": "mean"

}).round(2)

summary
```
