```python
# Step 1: Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots in notebook
%matplotlib inline

print("Libraries loaded successfully")
```

```python
# Step 2: Create sales and profit dataset

data = {
    "Company": [
        "Clas Ohlson", "Clas Ohlson", "Clas Ohlson", "Clas Ohlson",
        "Wickes", "Wickes", "Wickes", "Wickes"
    ],
    
    "Year": [
        2022, 2023, 2024, 2025,
        2022, 2023, 2024, 2025
    ],
    
    "Revenue_Million": [
        10500, 11000, 11600, 12030,   # Clas Ohlson SEK millions
        1562, 1553, 1538, 1590        # Wickes GBP millions
    ],
    
    "Gross_Margin_%": [
        38.5, 39.8, 40.2, 40.5,
        36.6, 36.4, 36.8, 37.0
    ],
    
    "Operating_Margin_%": [
        7.0, 8.5, 10.1, 10.3,
        4.4, 4.0, 4.7, 4.8
    ],
    
    "Net_Margin_%": [
        6.2, 7.5, 8.5, 8.8,
        2.0, 1.7, 2.5, 2.7
    ]
}

df_sales = pd.DataFrame(data)

df_sales
```

```python
# Step 3: Calculate profit values

df_sales["Gross_Profit"] = df_sales["Revenue_Million"] * df_sales["Gross_Margin_%"] / 100
df_sales["Operating_Profit"] = df_sales["Revenue_Million"] * df_sales["Operating_Margin_%"] / 100
df_sales["Net_Profit"] = df_sales["Revenue_Million"] * df_sales["Net_Margin_%"] / 100

df_sales
```

```python
# Step 4: Create demographics dataset

demographics = {
    "Company": ["Clas Ohlson"]*5 + ["Wickes"]*5,
    
    "Age_Group": [
        "18-24", "25-34", "35-44", "45-54", "55+",
        "18-24", "25-34", "35-44", "45-54", "55+"
    ],
    
    "Customer_Percentage": [
        12, 22, 28, 23, 15,   # Clas Ohlson
        8, 18, 30, 28, 16     # Wickes
    ],
    
    "Gender_Male_%": [
        48, 52, 55, 58, 60,
        60, 65, 70, 72, 75
    ],
    
    "Gender_Female_%": [
        52, 48, 45, 42, 40,
        40, 35, 30, 28, 25
    ]
}

df_demo = pd.DataFrame(demographics)

df_demo
```

```python
# Step 5: Save dataset

df_sales.to_csv("sales_profit_dataset.csv", index=False)
df_demo.to_csv("demographics_dataset.csv", index=False)

print("Datasets saved successfully")
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=df_sales,
    x="Year",
    y="Revenue_Million",
    hue="Company",
    marker="o"
)

plt.title("Revenue Comparison")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=df_sales,
    x="Year",
    y="Operating_Margin_%",
    hue="Company"
)

plt.title("Operating Margin Comparison")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=df_demo,
    x="Age_Group",
    y="Customer_Percentage",
    hue="Company"
)

plt.title("Customer Age Demographics")
plt.show()
```

```python
pivot = df_sales.pivot(
    index="Year",
    columns="Company",
    values="Net_Profit"
)

pivot.plot(kind="bar", figsize=(10,6))

plt.title("Net Profit Comparison")
plt.show()
```
